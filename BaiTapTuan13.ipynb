```python
#W11A1
def reverse(path: str) -> int:
    with open(path, "r") as f:
        n_str = f.read().strip()
    reversed_str = n_str[::-1]
    return int(reversed_str)
print(reverse("number.txt"))
```

```python
#W11A4
def areAnagrams(path1: str, path2: str) -> bool:
    with open(path1, "r", encoding="utf-8") as f:
        s1 = f.read()
    with open(path2, "r", encoding="utf-8") as f:
        s2 = f.read()
    def normalize(s: str):
        s = s.lower()
        letters = [ch for ch in s if ch.isalpha()]
        return sorted(letters)
    return normalize(s1) == normalize(s2)
print(areAnagrams("a.txt", "b.txt"))
```

```python
#W11A5
def count(path: str, character: str) -> int:
    with open(path, "r", encoding="utf-8") as f:
        sentence = f.read()
    return sentence.count(character)
print(count("number.txt", "l"))
```

```python
#W11A6
def countCharacter(path: str, letter: str) -> int:
    with open(path, "r", encoding="utf-8") as f:
        text = f.read()
    text = text.lower()
    letter = letter.lower()
    return text.count(letter)
print(countCharacter("text.txt", "e"))
```

```python
#W11A10
def rob(money):
    if not money:
        return 0
    if len(money) == 1:
        return money[0]
    dp0 = money[0]
    dp1 = max(money[0], money[1])
    for i in range(2, len(money)):
        dpi = max(dp1, dp0 + money[i])
        dp0 = dp1
        dp1 = dpi
    return dp1
money = list(map(int, input().split()))
print(rob(money))
```

```python
#W11A12
def maximumProduct(path: str) -> int:
    try:
        with open(path, "r") as f:
            lines = f.read().strip().split()
            n = int(lines[0])
            arr = list(map(int, lines[1:]))
            arr = arr[:n]
            arr.sort()
            prod1 = arr[-1] * arr[-2] * arr[-3]
            prod2 = arr[0] * arr[1] * arr[-1]
            return max(prod1, prod2)
    except:
        print("Mission failed")
        return None
path = input().strip()
print(maximumProduct(path))
```

```python
#W11A13
def moveZeroes(fileName: str) -> list:
    try:
        with open(fileName, "r") as f:
            lines = f.read().split()           
            n = int(lines[0])
            arr = list(map(int, lines[1:1+n]))            
            non_zero = [x for x in arr if x != 0]
            zeroes = [0] * (n - len(non_zero))        
            return non_zero + zeroes
    except:
        print("Mission failed")
        return None
print(moveZeroes("data.txt"))
```

```python
#W11A14
def averageTime(path: str):
    try:
        with open(path, "r", encoding="utf-8-sig") as f:
            lines = f.read().splitlines()
            lines = lines[1:]
            total = 0.0
            count = 0
            for line in lines:
                if not line.strip():
                    continue
                parts = line.split(",")
                if len(parts) != 2:
                    raise ValueError("Wrong CSV format")
                _, time = parts
                total += float(time)
                count += 1
            avg = total / count
            return round(avg, 2)
    except Exception as e:
        print("Mission failed")
        return None
path = input().strip()
print(averageTime(path))
```

```python
#W11A15
def findInterestingMovies(path: str):
    try:
        with open(path, "r", encoding="utf-8-sig") as f:
            lines = f.read().splitlines()
            lines = lines[1:]            
            result = []
            for line in lines:
                if not line.strip():
                    continue
                parts = [x.strip() for x in line.split(",")]
                _, movie, description, rating = parts        
                rating = float(rating)
                if rating >= 8 and description.lower() != "boring":
                    result.append(movie)
            if len(result) == 0:
                return "All are boring!"
            return result
    except:
        print("Mission failed")
        return None
print(findInterestingMovies("movies.csv"))
```

```python
#W11A16
import math
def printPerimeters(path: str):
    try:
        with open(path, "r", encoding="utf-8-sig") as f:
            lines = [line.strip() for line in f.readlines()]
        i = 0
        while i < len(lines):
            if lines[i] == "":
                i += 1
                continue
            shape_type = lines[i].split(":")[1].strip()
            params = lines[i + 1]
            if shape_type == "SQUARE":
                side = float(params.split(":")[1].strip())
                perimeter = 4 * side
            elif shape_type == "RECTANGLE":
                parts = params.replace(":", "").split()
                width = float(parts[1])
                height = float(parts[3])
                perimeter = 2 * (width + height)
            elif shape_type == "CIRCLE":
                radius = float(params.split(":")[1].strip())
                perimeter = 2 * math.pi * radius
            else:
                raise ValueError("Unknown shape")
            print(f"{perimeter:.2f}")
            i += 2
    except:
        print("Mission failed")
path = input().strip()
printPerimeters(path)
```

```python
#W11A17
def productExceptSelf(path: str) -> list[int]:
    try:
        with open(path, "r", encoding="utf-8-sig") as f:
            data = f.read().split()          
            n = int(data[0])
            arr = list(map(int, data[1:1+n]))
        result = [0] * n
        zero_count = arr.count(0)        
        if zero_count > 1:
            result = [0] * n       
        elif zero_count == 1:
            total = 1
            zero_index = arr.index(0)
            for x in arr:
                if x != 0:
                    total *= x
            result[zero_index] = total    
        else:
            total = 1
            for x in arr:
                total *= x
            for i in range(n):
                result[i] = total // arr[i]
        print(*result)
        return result
    except:
        print("Mission failed")
        return None
productExceptSelf("data.txt")
```

```python
#W11A18
import math
def estimatedTime(path: str):
    try:
        with open(path, "r", encoding="utf-8-sig") as f:
            lines = f.read().splitlines()
        src_parts = lines[0].replace(":", "").split()
        x1 = float(src_parts[2])
        y1 = float(src_parts[4])
        des_parts = lines[1].replace(":", "").split()
        x2 = float(des_parts[2])
        y2 = float(des_parts[4])
        vel_parts = lines[2].replace(":", "").split()
        v = float(vel_parts[1])
        distance = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
        time = distance / v
        return round(time, 2)
    except:
        print("Mission failed")
        return None
path = input().strip()
print(estimatedTime(path))
```

```python
#W11A20
def imageSmoother(path: str):
    try:
        with open(path, "r", encoding="utf-8-sig") as f:
            data = f.read().split()            
        M = int(data[0])
        nums = list(map(int, data[1:]))
        img = []
        idx = 0
        for _ in range(M):
            img.append(nums[idx:idx+M])
            idx += M
        res = [[0]*M for _ in range(M)]
        for i in range(M):
            for j in range(M):
                total = 0
                count = 0
                for x in range(i-1, i+2):
                    for y in range(j-1, j+2):
                        if 0 <= x < M and 0 <= y < M:
                            total += img[x][y]
                            count += 1
                res[i][j] = total // count
        for row in res:
            print(*row)
    except:
        print("Mission failed")
imageSmoother("image.txt")
```
