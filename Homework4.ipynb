```python
# W4A6
while True:
    a, b = map(int, input("Nhap hai so: ").split())
    if a <= b:
        print("Ban da nhap dung")
        break
    else:
        print("Vui long nhap so a <= b")
count = 0
for n in range(a, b + 1):
    if n < 2:
        songuyento = False
    else:
        songuyento = True
        for i in range(2, n):
            if n % i == 0:
                songuyento = False
    if songuyento == True:
        count += n
print(f'Tong cac so nguyen to trong doan [{a}, {b}] la: {count}')
```

```python
#W4A7
while True:
    n = int(input("Nhap so nguyen duong >= 2: "))
    if n >= 2:
        print("Ban da nhap dung yeu cau")
        break
    else:
        print("Vui long nhap lai theo dung yeu cau")
num = []
for i in range(1, n + 1):
    if n % i == 0:
        if i < 2:
            songuyento = False
        else:
            songuyento = True
            for j in range(2, i):
                if i % j == 0:
                    songuyento = False
        if songuyento == True:
            num.append(i)
print(f'Uoc so nguyen to lon nhat cua n la:', max(num))
```

```python
#W4A9
n = int(input("Nhap so: "))
num = []
i = 1
while i * i <= n:
    sq = i * i
    s = str(sq)
    if len(s) == len(set(s)):
        num.append(sq)
    i += 1
print(*num)
```

```python
#W4A10
n = int(input("Nhap so: "))
max_len = 0
best_x = 1
for x in range(1, n + 1):
    t = x
    length = 1
    while t != 1:
        if t % 2 == 0:
            t //= 2
        else:
            t = 3 * t + 1
        length += 1
    if length > max_len or (length == max_len and x < best_x):
        max_len = length
        best_x = x
print(best_x, max_len)
```
