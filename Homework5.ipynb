```python
#W5A3
def So_nguyen_to(a):
    if a<2:
        return False
    for i in range(2, int(a**0.5) + 1):
        if a % i == 0:
            return False
    return True
a = int(input("Nhap so:"))
if So_nguyen_to(a):
    print(a, "la so nguyen to")
else:
    print(a, "khong la so nguyen to")
```

```python
#W5A4
a = int(input("Nhap so:"))
def perfectnumber(a):
    if a<1:
        return False
    tong = 0
    for i in range(1,a):
        if a % i ==0:
            tong += i
    return tong == a
if perfectnumber(a):
    print(a, "la so hoan hao")
else:
    print(a, "khong la so hoan hao")
```

```python
#W5A5
lst = [1, 3, 6, 2, 2, 4]
k = 4
def findfirstindex(lst, k):
    for i in range(len(lst)):
        if lst[i] == k:
            return i
    return -1
index = findfirstindex(lst, k)
print("Vi tri dau tien co gia tri", k, "la:", index)
```

```python
#W5A7
def calculator(num1, num2, operate):
    if operate == '+':
        return num1 + num2
    elif operate == '-':
        return num1 - num2
    elif operate == '*':
        return num1 * num2
    elif operate == '/':
        if num2 == 0:
            raise ValueError("Khong the chia cho 0")
        return num1 / num2
    else:
        raise ValueError("Phep toan khong hop le")
num1 = float(input("So thu nhat: "))
num2 = float(input("So thu hai: "))
operate = input("Nhap phep toan (+, -, *, /): ")
try:
    result = calculator(num1, num2, operate)
    print("Ket qua phep toan la:", f'{result:.2f}')
except ValueError as e:
    print(e)
```

```python
#W5A8
def hamming_distance(x, y):
    xor = x ^ y
    distance = 0
    while xor:
        distance += xor & 1
        xor >>= 1
    return distance
x = int(input())
y = int(input())
print("Khoang cach hamming giua",x ,"va", y, "la", hamming_distance(x, y))
```

```python
#W5A9
def tongcacchuso(n):
    if n<0:
        raise ValueError("n phai la so nguyen duong")
    if n == 0:
        return 0
    total = 0
    while n>0:
        total += n%10
        n //= 10
    return total
try:
    n=int(input("Nhap so nguyen duong n:"))
    print("Tong cac chữ so cua", n, "la:", tongcacchuso(n))
except ValueError as e:
    print(e)
```
