```python
#W9A1
def merge_sorted_lists(a, b):
    i = j = 0
    result = []
    while i < len(a) and j < len(b):
        if a[i] <= b[j]:
            result.append(a[i])
            i += 1
        else:
            result.append(b[j])
            j += 1
    while i < len(a):
        result.append(a[i])
        i += 1
    while j < len(b):
        result.append(b[j])
        j += 1
    return result
a = list(map(int, input().split()))
b = list(map(int, input().split()))
print(*merge_sorted_lists(a, b))
```

```python
#W9A2
import heapq
matrix = []
while True:
    s = input().strip()
    if s == "":
        break
    matrix.append(list(map(int, s.split())))
k = int(matrix.pop()[0])  # dòng cuối là k
n = len(matrix)
m = len(matrix[0])
visited = set()
heap = []
heapq.heappush(heap, (matrix[0][0], 0, 0))
visited.add((0, 0))
count = 0
while heap:
    val, i, j = heapq.heappop(heap)
    count += 1
    if count == k:
        print(val)
        break
    if i + 1 < n and (i + 1, j) not in visited:
        heapq.heappush(heap, (matrix[i + 1][j], i + 1, j))
        visited.add((i + 1, j))
    if j + 1 < m and (i, j + 1) not in visited:
        heapq.heappush(heap, (matrix[i][j + 1], i, j + 1))
        visited.add((i, j + 1))
```

```python
#W9A3
s = input().strip()
n = len(s)
digits = list(s)
best = s
for i in range(n):
    for j in range(i + 1, n):
        # tránh số bắt đầu bằng 0
        if i == 0 and digits[j] == '0':
            continue
        tmp = digits[:]
        tmp[i], tmp[j] = tmp[j], tmp[i]
        candidate = ''.join(tmp)
        if candidate < best:
            best = candidate
print(best)
```

```python
#W9A4
Q = int(input())
for _ in range(Q):
    a = input().strip()
    b = input().strip()
    n, m = len(a), len(b)
    dp = [False] * (m + 1)
    dp[0] = True
    for ch in a:
        new_dp = dp[:]
        if ch.islower():
            for j in range(m, 0, -1):
                if ch.upper() == b[j - 1] and dp[j - 1]:
                    new_dp[j] = True
        else:
            new_dp = [False] * (m + 1)
            for j in range(m, 0, -1):
                if ch == b[j - 1] and dp[j - 1]:
                    new_dp[j] = True
        dp = new_dp
    print("YES" if dp[m] else "NO")
```

```python
#W9A5
def is_subarray(list1, list2):
    n = len(list1)
    m = len(list2)
    for i in range(m - n + 1):
        if list2[i:i+n] == list1:
            return True
    return False
list1 = list(map(int, input().split()))
list2 = list(map(int, input().split()))
print("YES" if is_subarray(list1, list2) else "NO")
```

```python
#W9A6
n = int(input())
a = list(map(int, input().split()))
i = n - 2
while i >= 0 and a[i] >= a[i + 1]:
    i -= 1
if i == -1:
    a.reverse()
else:
    j = n - 1
    while a[j] <= a[i]:
        j -= 1
    a[i], a[j] = a[j], a[i]
    a[i + 1:] = reversed(a[i + 1:])
print(*a)
```

```python
#W9A7
def canPlaceFlowers(flowerbed, k):
    n = len(flowerbed)
    count = 0
    for i in range(n):
        if flowerbed[i] == 0:
            left = (i == 0) or (flowerbed[i - 1] == 0)
            right = (i == n - 1) or (flowerbed[i + 1] == 0)
            if left and right:
                flowerbed[i] = 1
                count += 1
                if count >= k:
                    return True
    return count >= k
flowerbed = list(map(int, input().split()))
k = int(input())
print(canPlaceFlowers(flowerbed, k))
```

```python
#W9A8
def is_path_crossing(moves):
    x = y = 0
    visited = {(0, 0)}
    for move in moves:
        if move == 'R':
            x += 1
        elif move == 'L':
            x -= 1
        elif move == 'U':
            y += 1
        elif move == 'D':
            y -= 1
        if (x, y) in visited:
            return True
        visited.add((x, y))
    return False
moves = input().strip()
print(is_path_crossing(moves))
```

```python
#W9A9
def Median(nums):
    a, b, c, d, e = nums
    if a > b: a, b = b, a
    if c > d: c, d = d, c
    if a > c:
        a, c = c, a
        b, d = d, b
    if b > e: b, e = e, b
    if b > c: b, c = c, b
    if d < b: d = b
    if d > e: d = e
    return d
nums = map(int, input().split())
print(Median(nums))
```

```python
#W9A10
import bisect
def findRadius(houses, heaters):
    heaters.sort()
    ans = 0
    for h in houses:
        pos = bisect.bisect_left(heaters, h)
        left_dist = float('inf')
        right_dist = float('inf')
        if pos > 0:
            left_dist = h - heaters[pos - 1]
        if pos < len(heaters):
            right_dist = heaters[pos] - h
        ans = max(ans, min(left_dist, right_dist))
    return ans
houses = list(map(int, input().split()))
heaters = list(map(int, input().split()))
print(findRadius(houses, heaters))
```
