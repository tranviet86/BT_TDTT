```python
#W7A1
def binary_search(a, n):
    for i in range(len(a)):
        if a[i] == n:
            return i
    return "-1"
a = list(map(int, input().split()))
n = int(input())
print(binary_search(a, n))
```

```python
#W7A2
def count_occurrences(arr, x):
    count = 0
    for i in range(len(arr)):
        if arr[i] == x:
            count += 1
    return count
arr = list(map(int, input().split()))
x = int(input())
print(count_occurrences(arr, x))
```

```python
#W7A3
def sort(a):
    return sorted(a)
a = list(map(int, input().split()))
print(*sort(a))
```

```python
#W7A4
def solan(a):
    max_count = 0
    result = a[0]
    for i in a:
        count = a.count(i)
        if count > max_count:
            max_count = count
            result = i
    return result, max_count
a = list(map(int, input().split()))
values, time = solan(a)
print(f'{values} xuat hien nhieu nhat, som nhat, {time} lan')
```

```python
#W7A5
a = list(map(int, input().split()))
x = int(input())
count = 0
for i in range(len(a)):
    for j in range(i + 1, len(a)):
        if a[i] + a[j] == x:
            count += 1
print(count)
```

```python
#W7A7
nums = eval(input())
n = len(nums)
dp = [1] * n
for i in range(n):
    for j in range(i):
        if nums[j] < nums[i]:
            dp[i] = max(dp[i], dp[j] + 1)
print(max(dp))
```

```python
#W7A9
a = list(map(int, input().split()))
b = []
for i in range(len(a)):
    count = 0
    for j in range(i + 1, len(a)):
        if a[i] > a[j]:
            count += 1
    b.append(count)
print(b)
```

```python
#W7A10
lst = eval(input())
n = int(input())
ketqua = []
for i in range(len(lst)):
    for j in range(i + 1, len(lst) + 1):
        a = lst[i:j]
        b = sum(a) % n
        ketqua.append(b)
print(max(ketqua))
```

```python
#W7A11
def timkiem(a, b):
    count = 0
    for i in a:
        if i == b:
            count += 1
    return count
a = input()
a = a.replace(',', '').replace('.', '')
a = a.split()
b = input()
print(timkiem(a, b))
```

```python
#W7A12
words = []
labels = ['A', 'B', 'C', 'D']
for i in range(4):
    s = input().strip()
    words.append((s, labels[i]))
words.sort(key = lambda x: x[0].lower())
for word, labels in words:
    print(labels, end =' ')
```

```python
#W7A13
n = list(map(int, input().split()))
flat =[]
for i in range(len(n)):
    count = 0
    for j in range(len(n)):
        if n[j] == n[i] + 1 or n[i] == n[j]:
            count += 1
    flat.append(count)
print(max(flat))
```
