```python
#Bai4
h = int(input("Nhap do cao: "))
for i in range(1, h + 1):
    print(' ' * (h - i), end = '')
    if i == 1:
        print('*')
    elif i == h:
        print('*' * (2 * h - 1))
    else:
        print('*' + ' ' * (2 * i - 3) + '*')
```

```python
#Bai9
a = int(input("Nhap so: "))
i = 1
while i * i <= a:
    if i * i == a:
        print(a, "la so chinh phuong")
        break
    i += 1
else:
    print(a, "khong phai so chinh phuong")
```

```python
#Bai15
n = int(input("Nhap so: "))
count = 0 
if n == 0:
    count = 1
else:
    for i in range(1, len(str(abs(n))) + 1):
        count += 1
print(n, "co", count, "chu so")
```

```python
#Bai16
n = int(input("Nhap so: "))
m = n
even = 0
odd = 0
while n > 0:
    a = n % 10
    if a % 2 == 0:
        even += 1
    else:
        odd += 1
    n //= 10
print(m, "co", even, "chu so chan va", odd, "chu so le")
```

```python
#Bai17
n = int(input("Nhap so: "))
m = n
count = 0
while n > 0:
    count = count + (n % 10)
    n //= 10
    if n == 0:
        break
print("Tong cac chu so cua ", m, "la", count)
```

```python
#Bai18
n = int(input("Nhap so: "))
for i in range(1, 100):
    if n == 3 ** i:
        print(n, "la so co dang 3^k")
        break
else:
    print(n, "khong co dang 3^k")
```

```python
#Bai20
n = int(input("Nhap so: "))
m = n
songuyento = False
count = 0
if n < 2:
    songuyento = False
else:
    for i in range(2, n):
        a = n % 10
        if a % i == 0:
            songuyento = False
        else:
            songuyento = True
            count += 1
        n //= 10
print(f' {m} co {count} so nguyen to')
```
