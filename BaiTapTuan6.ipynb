```python
#W6A4
line = input().strip()
pairs = line.split()
result = {}
for p in pairs:
    key, value = p.split(":")
    if key not in result:
        result[key] = []
    result[key]. append(value)
print(result)
```

```python
#W6A5
lst = map(int, input("List so nguyen").split())
dic = {
    'positive': 0,
    'negative': 0,
    'zeros': 0
}
for x in lst:
    if x > 0:
        dic['positive'] += 1
    elif x < 0:
        dic['negative'] += 1
    else:
        dic['zeros'] += 1
print(dic)
```

```python
#W6A9
d1 = eval(input())
d2 = eval(input())
result = {}
for x in d1:
    result[x] = d1[x]
for x in d2:
    if x in result:
        result[x] += d2[x]
    else:
        result[x] = d2[x]
print(result)
```

```python
#W6A10
lst = list(map(int, input("Nhap danh sach: ").split()))
k = int(input("Nhap so: "))
res = []
for i in range(len(lst)):
    for j in range(i + 1, len(lst)):
        if lst[i] + lst[j] == k:
            res.append((lst[i], lst[j]))
print(res)
```

```python
#W6A12
nums = list(map(int, input().split()))
count = {}
for x in nums:
    if x in count:
        count[x] += 1
    else:
        count[x] = 1
max_freq = max(count.values())
candidates = []
for key in count:
    if count[key] == max_freq:
        candidates.append(key)
print(min(candidates))
```

```python
#W6A14
lst1 = list(map(int, input("Danh sach 1: ").split()))
lst2 = list(map(int, input("Danh sach 2: ").split()))
res = []
seen = set()
for num in lst1:
    if num in lst2 and num not in seen:
        res.append(num)
        seen.add(num)
print(res)
```
