```python
#W8A2
class Date:
    def __init__(self, d, m, y):
        self.d = d
        self.m = m
        self.y = y
    def NamNhuan(self):
        return (self.y % 400 == 0) or (self.y % 4 == 0 and self.y % 100 != 0)
    def invalid(self):
        if self.m < 1 or self.m > 12:
            return False
        if self.m in [1, 3, 5, 7, 8, 10, 12]:
            max_day = 31
        elif self.m in [4, 6, 9, 11]:
            max_day = 30
        else:
            if self.NamNhuan():
                max_day = 29
            else:
                max_day = 28
        if self.d < 1 or self.d > max_day:
            return False
        return True
    def next_day(self):
        self.d += 1
        if self.m in [1, 3, 5, 7, 8, 10, 12]:
            max_day = 31
        elif self.m in [4, 6, 9, 11]:
            max_day = 30
        else:
            if self.NamNhuan():
                max_day = 29
            else:
                max_day = 28
        if self.d > max_day:
            self.d = 1
            self.m += 1
            if self.m > 12:
                self.m = 1
                self.y += 1
        return self.d, self.m, self.y 
s = input().strip()
d, m, y = map(int, s.split('/'))
date = Date(d, m, y)
if not date.invalid():
    print("INVALID")
else:
    nd, nm, ny = date.next_day()
    print(f'{nd:02d}/{nm:02d}/{ny:02d}')
```

```python
#W8A3
import math
class lop_diem:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    def diem_M(self):
        if self.x == 0 and self.y == 0:
            return "M la goc toa do"
        elif self.x == 0 and self.y != 0:
            return "M nam tren truc tung"
        elif self.x != 0 and self.y == 0:
            return "M nam tren truc hoanh"
        else:
            return "M la diem bat ki"
    def distance(self):
        return math.sqrt(self.x * self.x + self.y * self.y)
s = input().strip()
s = s.replace('(', '').replace(',', '').replace(')', '')
x, y = map(float, s.split())
a = lop_diem(x, y)
diemM = a.diem_M()
dis = a.distance()
print(f'{diemM}, khoang cach tu goc toa do toi diem M la: {dis:.2f}')
```

```python
#W8A4
class Calculator:
    def __init__(self, num1, num2):
        self.num1 = num1
        self.num2 = num2
    def add(self):
        return self.num1 + self.num2
    def subtract(self):
        return self.num1 - self.num2
    def product(self):
        return self.num1 * self.num2
    def divide(self):
        if self.num2 == 0:
            return "INVALID"
        else:
            return self.num1 / self.num2
    def power(self):
        return self.num1 ** self.num2
    def mod(self):
        if self.num2 == 0:
            return "INVALID"
        else:
            return self.num1 % self.num2
    def set_numbers(self, num1, num2):
        self.num1 = num1
        self.num2 = num2 
num1, num2 = map(float, input("Nhap hai so can tinh: ").split())
cal = Calculator(num1, num2)
while True:
    print("\n==== MENU ====")
    print("1. Cộng (add)")
    print("2. Trừ (subtract)")
    print("3. Nhân (product)")
    print("4. Chia (divide)")
    print("5. Mũ (power)")
    print("6. Phần dư (mod)")
    print("7. Cập nhật số mới (set_numbers)")
    choice = input("Chọn phép toán từ 1-7: ")
    if choice == "1":
        print("Kết quả: ", cal.add())
    elif choice == "2":
        print("Kết quả: ", cal.subtract())
    elif choice == "3":
        print("Kết quả: ", cal.product())
    elif choice == "4":
        print("Kết quả: ", cal.divide())
    elif choice == "5":
        print("Kết quả: ", cal.power())
    elif choice == "6":
        print("Kết quả: ", cal.mod())
    elif choice == "7":
        num1, num2 = map(float, input("Nhập hai số mới: ").split())
        cal.set_numbers(num1, num2)
        print("Bạn đã cập nhật 2 số mới")
    else:
        print("Lựa chọn không hợp lệ")
    exit_choice = input("Bạn có muốn thoát?, (Nhập YES/Yes/yes để thoát")
    if exit_choice in ["yes", "YES", "Yes"]:
        print("Bạn đã thoát chương trình")
        break
```

```python
#W8A5
class ShoppingCart:
    def __init__(self):
        self.products = []
    def add_product(self, name, price):
        product = {"name": name, "price": price}
        self.products.append(product)
        print(f'Da them san pham: {name} - {price}VND')
    def remove_product(self, name):
        if product["name"].lower() == name.lower():
            self.products.remove(product)
            print(f"Da xoa san pham: {name} - {price}VND")
    def is_empty(self):
        return len(self.products) == 0
    def total_price(self):
        total = sum(product["price"] for product in self.products)
        return total
    def show_products(self):
        if self.is_empty():
            print("Gio hang rong")
        else:
            print("Danh sach san pham trong gio hang")
            i = 1
            for product in self.products:
                print(f'{i}. {product["name"]} - {product["price"]}VND')
                i += 1
            print(f"Tong tien la: {self.total_price()}")
    def clear_products(self):
        self.products.clear()
        print("Da xoa toan bo san pham trong gio hang")
cart = ShoppingCart()
while True:
    print("\n---- MENU GIO HANG ----")
    print("2. Them san pham vao gio")
    print("3. Xoa san pham khoi gio (theo ten)")
    print("4. Kiem tra gio hang")
    print("5. Tinh tong tien gio hang")
    print("6.Hien thi toan bo san pham trong gio")
    print("7. Xoa toan bo gio hang")
    choice = input("Chon chuc nang tu 2-7")
    if choice == "2":
        name = input("Nhap ten san pham: ")
        price = float(input("Nhap gia san pham: "))
        cart.add_product(name, price)
    elif choice == "3":
        cart.remove_product(name)
    elif choice == "4":
        if cart.is_empty():
            print("Gio hang dang trong")
        else:
            print("Gio hang dang co san pham")
    elif choice == "5":
        print(f'Tong tien la: {cart.total_price()}VND')
    elif choice == "6":
        cart.show_products()
    elif choice == "7":
        cart.clear_products()
    else:
        print("Vui long chon chuc nang tu 2-7")
    exit_choice = input("Co muon thoat chuc nang khong? (yes/no)")
    if exit_choice == "yes":
        print("Ban da thoat chuc nang")
        break
```

```python
#W8A6
class Rectangle:
    def __init__(self, w, h, k):
        self.w = w
        self.h = h
        self.k = k
    def area(self):
        return self.w * self.h
    def perimeter(self):
        return (self.w + self.h) * 2
    def scale(self):
        area = self.area() * (self.k ** 2)
        perimeter = self.perimeter() * self.k
        return area, perimeter
w, h, k = map(float, input().split())
rec = Rectangle(w, h, k)
new_are, new_perimeter = rec.scale()
print(f'{new_are:.2f} {new_perimeter:.2f}')
```

```python
#W8A7
import math
class Fraction:
    def __init__(self, num, den):
        if den == 0:
            raise ValueError("Mau so khong the bang 0")
        if den < 0:
            num = -num
            den = -den
        g = math.gcd(num, den)
        self.num = num // g
        self.den = den // g
    def add(self, other):
        num = self.num * other.den + other.num * self.den
        den = self.den * other.den
        return Fraction(num, den)
    def sub(self, other):
        num = self.num * other.den - other.num * self.den
        den = self.den * other.den
        return Fraction(num, den)
    def mul(self, other):
        num = self.num * other.num
        den = self.den * other.den
        return Fraction(num, den)
    def div(self, other):
        num = self.num * other.den
        den = self.den * other.num
        return Fraction(num, den)
    def __str__(self):
        return f'{self.num}/{self.den}'
a, b, op, c, d = input().split()
a, b, c, d = int(a), int(b), int(c), int(d)
f1 = Fraction(a, b)
f2 = Fraction(c, d)
if op == '+':
    result = f1.add(f2)
elif op == '-':
    result = f1.sub(f2)
elif op == '*':
    result = f1.mul(f2)
else:
    result = f1.div(f2)
print(result)
```

```python
#W8A8
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = float(balance)
    def deposit(self, x):
        self.balance += x
    def withdraw(self, x):
        if x <= self.balance:
            self.balance -= x
    def __str__(self):
        return f'{self.owner}: {self.balance:.2f}'
owner, balance = input().split()
balance = float(balance)
account = BankAccount(owner, balance)
n = int(input())
for _ in range(n):
    cmd, x = input().split()
    x = float(x)
    if cmd == "DEPOSIT":
        account.deposit(x)
    elif cmd == "WITHDRAW":
        account.withdraw(x)
print(f'{account.balance:.2f}')
```

```python
#W8A13
class Triangle:
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c
    def getArea(self):
        if self.a + self.b > self.c and self.a + self.c > self.b and self.b + self.c > self.a:
            p = (self.a + self.b + self.c) / 2
            return math.sqrt(p * (p - self.a) * (p - self.b) * (p - self.c))
        else:
            return "invalid"
a, b, c = map(int, input().split())
trig = Triangle(a, b, c)
x = trig.getArea()
if x == "invalid":
    print("invalid")
else:
    print(f'{x:.2f}')
```
