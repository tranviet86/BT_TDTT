```python
#W6A16
n, m = map(int, input().split())
a =[]
for i in range(n):
    row = list(map(int, input().split()))
    a.append(row)
for i in range(n):
    for j in range(m):
        print(f'{a[i][j]:4d}', end ='')
    print()
```

```python
#W6A17
n = int(input("Ma tran co: "))
a =[]
for i in range(n):
    row = list(map(int, input().split()))
    a.append(row)
print("Duong cheo chinh la: ")
for i in range(n):
    print(f'{a[i][i]:4d}', end ='')
print()
print("Duong cheo phu la: ")
for i in range(n):
    print(f'{a[i][n-1-i]:4d}', end ='')
```

```python
#W6A18
n, m, k = map(int, input().split())
a = []
for i in range(n):
    row = list(map(int, input().split()))
    a.append(row)
tong = 0
for j in range(m):
    tong += a[k][j]
for i in range(n):
    if i != k:
        tong += a[i][k]
print(tong)
```

```python
#W6A20
def songuyen(n,k):
    for i in range(len(n)):
        if n[i] == k:
            return i
            break
    return "-1"
n = list(map(int, input().split()))
k = int(input())
print(songuyen(n,k))
```

```python
#W6A23
def xoaso(x,a):
    b = []
    for i in a:
        if i >= x:
            b.append(i)
    return b
a = list(map(int, input().split()))
x = int(input())
print(xoaso(x, a))
```

```python
#W6A24
def sapxep(a):
    b = []
    for i in a:
        if i == 0:
            b.append(i)
    for i in a:
        if i != 0:
            b.append(i)
    return b
a = list(map(int, input().split()))
print(*sapxep(a))
```

```python
#W6A26
a = list(map(int, input().split()))
s = 0
for i in range(1, len(a) - 1):
    if a[i] > a[i - 1] and a[i] > a[i + 1]:
        s += a[i]
print(s)
```

```python
#W6A28
a = list(map(int, input().split()))
tang = True
giam = True
for i in range(len(a) - 1):
    if a[i] >= a[i + 1]:
        tang = False
    if a[i] <= a[i + 1]:
        giam = False
if tang:
    print("Day tang thuc su")
elif giam:
    print("Day giam thuc su")
else:
    print("Vo danh")
```
