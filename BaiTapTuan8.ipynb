```python
#W7A16
def mang_dep(a):
    dem = {}
    for x in a:
        if x in dem:
            dem[x] += 1
        else:
            dem[x] = 1
    xoa = 0
    for x in dem:
        if dem[x] < x:
            xoa += dem[x]
        elif dem[x] > x:
            xoa += dem[x] - x
    return xoa
a = list(map(int, input().split()))
if mang_dep(a) == 0:
    print('YES')
else:
    print('NO', mang_dep(a))
```

```python
#W7A18
n = list(map(int, input().split()))
even = []
odd = []
for i in n:
    if i % 2 == 0:
        even.append(i)
    else:
        odd.append(i)
even.sort()
odd.sort()
odd.reverse()
a = even + odd
print(*a)
```

```python
#W7A19
def sap_xep(a, b):
    i = j = 0
    c = []
    increasing = True
    if len(a) > 1 and a[0] > a[-1]:
        increasing = False
    if increasing:
        while i < len(a) and j < len(b):
            if a[i] <= b[j]:
                c.append(a[i])
                i += 1
            else:
                c.append(b[j])
                j += 1
    else:
        while i < len(a) and j < len(b):
            if a[i] >= b[j]:
                c.append(a[i])
                i += 1
            else:
                c.append(b[j])
                j += 1
    while i < len(a):
        c.append(a[i])
        i += 1
    while j < len(b):
        c.append(b[j])
        j += 1
    return c
a = list(map(int, input().split()))
b = list(map(int, input().split()))
print(*sap_xep(a, b))
```

```python
#W7A20
def tinh_tong(a, b):
    tong = 0
    for i in range(len(a)):
        count = 0
        x = a[i]
        while x > 0:
            count += 1
            x //= 10
        if count == b:
            tong += a[i]
    return tong
a = list(map(int, input().split()))
b = int(input())
print(tinh_tong(a, b))
```

```python
#W7A21
N, T = map(int, input().split())
days_in_month = [31, 29, 31, 30, 31, 30,
                 31, 31, 30, 31, 30, 31]
if T < 1 or T > 12 or N < 1 or N > days_in_month[T - 1]:
    print("invalid")
else:
    day_of_year = sum(days_in_month[:T - 1]) + N
    print(day_of_year)
```

```python
#W7A22
n = int(input())
count = {}
for _ in range(n):
    full_name = input().strip()
    name = full_name.split()[-1]
    if name in count:
        count[name] += 1
    else:
        count[name] = 1
max_freg = max(count.values())
candidate = []
for name in count:
    if count[name] == max_freg:
        candidate.append(name)
result = min(candidate)
print(result)
```

```python
#W7A29
n = list(map(int, input().split()))
ds1 = [n[0]]
ds2 = []
flag = False
for i in range(1, len(n)):
    if n[i - 1] < n[i] and not flag:
        ds1.append(n[i])
    else:
        flag = True
        ds2.append(n[i])
if len(ds1) < len(ds2):
    print(*ds2)
elif len(ds1) > len(ds2):
    print(*ds1)
else:
    print(ds1)
```

```python
#W7A30
a = list(map(int, input().split()))
n = len(a)
dp = [1] * n
for i in range(n):
    for j in range(i):
        if a[j] < a[i]:
            dp[i] = max(dp[i], dp[j] + 1)
print(max(dp))
```
